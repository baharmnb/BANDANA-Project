```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load dataset
# -----------------------------
df = pd.read_csv("ai_job_dataset.csv")

# -----------------------------
# 2. Basic cleaning
# -----------------------------
df = df.dropna(subset=["salary_usd", "required_skills"]).copy()
df["salary_usd"] = pd.to_numeric(df["salary_usd"], errors="coerce")
df = df.dropna(subset=["salary_usd"])

# -----------------------------
# 3. Define AI-specific skills
# -----------------------------
# You can extend or edit this list based on your project framing
AI_SKILLS = {
    "Machine Learning",
    "Deep Learning",
    "NLP",
    "Computer Vision",
    "MLOps",
    "LLMs",
    "Generative AI",
    "Reinforcement Learning",
    "Neural Networks",
    "AI Ethics",
    "Prompt Engineering"
}

# -----------------------------
# 4. Split and explode skills
# -----------------------------
df["skill_list"] = (
    df["required_skills"]
    .astype(str)
    .str.split(",")
    .apply(lambda skills: [s.strip() for s in skills if s.strip()])
)

skills_df = df.explode("skill_list").rename(columns={"skill_list": "skill"})
skills_df["skill"] = skills_df["skill"].str.strip()

# -----------------------------
# 5. Classify each skill
# -----------------------------
skills_df["skill_group"] = skills_df["skill"].apply(
    lambda x: "AI-specific" if x in AI_SKILLS else "Supporting"
)

# -----------------------------
# 6. Overall salary baseline
# -----------------------------
overall_mean_salary = skills_df["salary_usd"].mean()
overall_median_salary = skills_df["salary_usd"].median()

# -----------------------------
# 7. Salary stats per skill
# -----------------------------
skill_salary_stats = (
    skills_df.groupby(["skill_group", "skill"])
    .agg(
        job_count=("salary_usd", "count"),
        mean_salary_usd=("salary_usd", "mean"),
        median_salary_usd=("salary_usd", "median"),
        std_salary_usd=("salary_usd", "std")
    )
    .reset_index()
)

skill_salary_stats["mean_salary_premium_vs_dataset"] = (
    skill_salary_stats["mean_salary_usd"] - overall_mean_salary
)

skill_salary_stats["median_salary_premium_vs_dataset"] = (
    skill_salary_stats["median_salary_usd"] - overall_median_salary
)

# -----------------------------
# 8. Filter rare skills
# -----------------------------
MIN_JOB_COUNT = 30
skill_salary_filtered = skill_salary_stats[
    skill_salary_stats["job_count"] >= MIN_JOB_COUNT
].copy()

# -----------------------------
# 9. Separate outputs
# -----------------------------
ai_skill_results = (
    skill_salary_filtered[skill_salary_filtered["skill_group"] == "AI-specific"]
    .sort_values(by="median_salary_usd", ascending=False)
)

supporting_skill_results = (
    skill_salary_filtered[skill_salary_filtered["skill_group"] == "Supporting"]
    .sort_values(by="median_salary_usd", ascending=False)
)

# -----------------------------
# 10. Print results
# -----------------------------
print("Overall mean salary in dataset: ${:,.2f}".format(overall_mean_salary))
print("Overall median salary in dataset: ${:,.2f}".format(overall_median_salary))

print("\nTop AI-specific skills by median salary:")
print(
    ai_skill_results[
        ["skill", "job_count", "mean_salary_usd", "median_salary_usd", "median_salary_premium_vs_dataset"]
    ].head(15)
)

print("\nTop supporting skills by median salary:")
print(
    supporting_skill_results[
        ["skill", "job_count", "mean_salary_usd", "median_salary_usd", "median_salary_premium_vs_dataset"]
    ].head(15)
)

# -----------------------------
# 11. Save results
# -----------------------------
ai_skill_results.to_csv("q1_ai_specific_skill_salary.csv", index=False)
supporting_skill_results.to_csv("q1_supporting_skill_salary.csv", index=False)
skill_salary_filtered.to_csv("q1_all_skill_salary_split.csv", index=False)

# -----------------------------
# 12. Plot top AI-specific skills
# -----------------------------
top_ai = ai_skill_results.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_ai["skill"], top_ai["median_salary_usd"])
plt.xlabel("Median Salary (USD)")
plt.ylabel("AI-specific Skill")
plt.title("Top 10 AI-specific Skills by Median Salary")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# -----------------------------
# 13. Plot top supporting skills
# -----------------------------
top_supporting = supporting_skill_results.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_supporting["skill"], top_supporting["median_salary_usd"])
plt.xlabel("Median Salary (USD)")
plt.ylabel("Supporting Skill")
plt.title("Top 10 Supporting Skills by Median Salary")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
