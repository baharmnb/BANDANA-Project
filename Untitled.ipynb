```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. Load dataset
# -----------------------------
df = pd.read_csv("ai_job_dataset.csv")

# -----------------------------
# 2. Basic cleaning
# -----------------------------
# Keep only rows with valid salary and skills
df = df.dropna(subset=["salary_usd", "required_skills"]).copy()

# Make sure salary is numeric
df["salary_usd"] = pd.to_numeric(df["salary_usd"], errors="coerce")
df = df.dropna(subset=["salary_usd"])

# -----------------------------
# 3. Split skills into lists
# -----------------------------
# Example: "Python, SQL, TensorFlow" -> ["Python", "SQL", "TensorFlow"]
df["skill_list"] = (
    df["required_skills"]
    .astype(str)
    .str.split(",")
    .apply(lambda skills: [s.strip() for s in skills if s.strip()])
)

# -----------------------------
# 4. Explode so each row = one job-skill pair
# -----------------------------
skills_df = df.explode("skill_list").rename(columns={"skill_list": "skill"})

# Optional: normalize skill naming
skills_df["skill"] = skills_df["skill"].str.strip()

# -----------------------------
# 5. Compute salary stats per skill
# -----------------------------
overall_mean_salary = skills_df["salary_usd"].mean()
overall_median_salary = skills_df["salary_usd"].median()

skill_salary_stats = (
    skills_df.groupby("skill")
    .agg(
        job_count=("job_id", "count"),
        mean_salary_usd=("salary_usd", "mean"),
        median_salary_usd=("salary_usd", "median"),
        std_salary_usd=("salary_usd", "std"),
        min_salary_usd=("salary_usd", "min"),
        max_salary_usd=("salary_usd", "max")
    )
    .reset_index()
)

# -----------------------------
# 6. Add salary premium columns
# -----------------------------
skill_salary_stats["mean_salary_premium_vs_dataset"] = (
    skill_salary_stats["mean_salary_usd"] - overall_mean_salary
)

skill_salary_stats["median_salary_premium_vs_dataset"] = (
    skill_salary_stats["median_salary_usd"] - overall_median_salary
)

# -----------------------------
# 7. Filter out very rare skills
# -----------------------------
# You can change this threshold depending on how strict you want to be
MIN_JOB_COUNT = 30

skill_salary_filtered = skill_salary_stats[
    skill_salary_stats["job_count"] >= MIN_JOB_COUNT
].copy()

# -----------------------------
# 8. Rank skills by salary
# -----------------------------
top_skills_by_mean = skill_salary_filtered.sort_values(
    by="mean_salary_usd", ascending=False
)

top_skills_by_median = skill_salary_filtered.sort_values(
    by="median_salary_usd", ascending=False
)

top_skills_by_premium = skill_salary_filtered.sort_values(
    by="mean_salary_premium_vs_dataset", ascending=False
)

# -----------------------------
# 9. Show results
# -----------------------------
print("Overall mean salary in dataset: ${:,.2f}".format(overall_mean_salary))
print("Overall median salary in dataset: ${:,.2f}".format(overall_median_salary))

print("\nTop 15 skills by MEAN salary:")
print(
    top_skills_by_mean[
        ["skill", "job_count", "mean_salary_usd", "median_salary_usd", "mean_salary_premium_vs_dataset"]
    ].head(15)
)

print("\nTop 15 skills by MEDIAN salary:")
print(
    top_skills_by_median[
        ["skill", "job_count", "mean_salary_usd", "median_salary_usd", "median_salary_premium_vs_dataset"]
    ].head(15)
)

print("\nTop 15 skills by salary premium vs dataset mean:")
print(
    top_skills_by_premium[
        ["skill", "job_count", "mean_salary_usd", "median_salary_usd", "mean_salary_premium_vs_dataset"]
    ].head(15)
)

# -----------------------------
# 10. Save result if needed
# -----------------------------
skill_salary_filtered.to_csv("q1_skill_salary_analysis.csv", index=False)
```
